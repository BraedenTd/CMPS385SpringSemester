```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_sal = pd.read_csv('/content/sample_data/insurance.csv')
df_sal.head()
```

```python
df_sal.describe()
```

```python
plt.title('Insurance Distribution Plot')
sns.distplot(df_sal['age'])
plt.show()
```

```python
plt.scatter(df_sal['bmi'], df_sal['age'], color = 'lightcoral')
plt.title('Age vs BMI')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.box(False)
plt.show()
```

```python
X = df_sal.iloc[:, :1]  # independent
y = df_sal.iloc[:, 1:]  # dependent
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Assuming 'sex' is the categorical column at index 0 of y
X = df_sal[['bmi','charges']]  # independent variables
y = df_sal['age']  # dependent variable

# Convert categorical features using one-hot encoding
X = pd.get_dummies(X, columns=['charges'], drop_first=True)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize and train the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
plt.scatter(X_train['bmi'], y_train, color = 'lightcoral') # Changed X_train to X_train['bmi']
plt.plot(X_train['bmi'], y_pred_train, color = 'firebrick') # Changed X_train to X_train['bmi']
plt.title('Age vs BMI (Training Set)')
plt.xlabel('BMI')
plt.ylabel('Age') # This ylabel might be incorrect, consider changing to 'Charges' if it represents the target variable.
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Age/BMI', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
